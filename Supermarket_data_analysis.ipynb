```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
df=pd.read_csv('/content/drive/MyDrive/supermarket sales analysis /supermarket_sales - Sheet1.csv')
df.head()
```

```python
''' barplot of gender '''
gender_label = df.Gender.value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_label.index,y= gender_label, alpha=0.8);
plt.xlabel('Gender', fontsize=20)
plt.ylabel('Count', fontsize=20);
```

```python
#one hot encoding
df['Gender']=np.where(df['Gender']=='Male',1,0)
df.head()
```

```python
#barplot of city 
city_label=df.City.value_counts()
plt.figure(figsize=(8,5))
sns.barplot(x=city_label.index,y=city_label);
plt.xlabel('City',fontsize=20)
plt.ylabel('Count', fontsize=20)
```

```python
#countplot
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sns.countplot(x='Product line', data=df, hue='Gender')
plt.xlabel('Product Line', fontsize=20)
plt.ylabel('Count',fontsize=20)
```

```python
#piechart
plt.figure(figsize=(10, 5))
plt.pie(df['Payment'].value_counts(),labels=df['Payment'].value_counts().keys(),autopct='%1.0f%%',radius=1.5)
plt.legend()
plt.show()
```

```python
#barplot of product line
product_line_label = df['Product line'].value_counts()
plt.figure(figsize=(10, 5))
plt.xticks(rotation=90)
sns.barplot(x=product_line_label.index,y= product_line_label);
plt.xlabel('Product Line', fontsize=20)
plt.ylabel('Count', fontsize=20);
```

```python
#value counts of quantity 
quantity_label = df.Quantity.value_counts()
print(quantity_label)

''' barplot of quantity '''
plt.figure(figsize=(8,6))
sns.barplot(x=quantity_label.index,y= quantity_label)
plt.xlabel('Quantity', fontsize=20)
plt.ylabel('Count', fontsize=20);
```

```python
#value counts of payment 
payment_label=df['Payment'].value_counts()
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.barplot(x=payment_label.index,y=payment_label);
plt.xlabel('Payment',fontsize=20)
plt.ylabel('Count',fontsize=20);
```

```python
#barplot
plt.figure(figsize=(8,6))
plt.xticks(rotation=90)
sns.barplot(x='Product line',y='gross income',data=df)
plt.xlabel('Product Line',fontsize=20)
plt.ylabel('Gross income',fontsize=20);
```
